```python
import pandas as pd
df=pd.read_csv("finaltrain.csv")
df.head()
```

```python
Dict={
         'Modèle': [],
         'État': [],
         'Année-Modèle': [],
        'Marque': [],
        'Origine': [],
        'Kilométrage':[],
        'Première main':[],
        'Nombre de portes':[]
    }
```

```python
def split_array(array):
    n=len(array)
    Title=[]
    Values=[]
    for i in range(len(array)):
        if i%2==0:
            Title.append(array[i])
        else: 
            Values.append(array[i])
    myDict = {}
    
    for i in range(len(Title)):
        myDict[Title[i]] = Values[i]
    return myDict
```

```python
Array=[i.split(':')[1][:-2][1:] for i in df["group one"][76].split(",")[3:]]
split_array(Array)
```

```python
import numpy as np
df1 = pd.DataFrame(Dict)
columns = ['Année-Modèle','Première main','Kilométrage', 'Nombre de portes','Origine','État','Marque','Modèle']

# Create a dictionary row
for index,row in df.iterrows():
    array= [i.split(':')[1][:-2][1:] for i in df.loc[index,"group one"].split(",")[3:]]
    myDict=split_array(array)
    newRow = pd.DataFrame.from_dict(myDict, orient='index').T
    newRow = newRow.reindex(columns=columns)
    df1 = pd.concat([df1, newRow], ignore_index=True)
# Create a DataFrame from the dictionary row


# Replace inexistent values with NaN

# Concatenate the new DataFrame with the original DataFrame

```

```python
import math
def swap_puissance(df):
    L=['carburant', 'Ville','boite', 'puissance']
    a=True
    for index, row in df.iterrows():
        for i in L:
            try:
                if isinstance(df.loc[index, i],str) and "CV" in df.loc[index, i] :
                    a=False
                    df.loc[index, "puissance"],df.loc[index, i]=df.loc[index, i],df.loc[index,"puissance"]
                    break;
            except ValueError:
                pass
        if a:
            for j in L:
                if isinstance(df.loc[index, j], float) :
                    if math.isnan(df.loc[index, j]):
                        df.loc[index, "puissance"],df.loc[index, j]=df.loc[index, j],df.loc[index, "puissance"]
        a=True
        
swap_puissance(data)
data.head()
```

```python
def swap_boite(df):
    L=['carburant', 'Ville','boite', 'puissance']
    a=True
    for index, row in df.iterrows():
        for i in L:
            if df.loc[index, i] in ["Automatique","Manuelle"]:
                a=False
                df.loc[index, "boite"],df.loc[index, i]=df.loc[index, i],df.loc[index,"boite"]
                break;
        
        if a:
            for j in L:
                if isinstance(df.loc[index, j], float) :
                    if math.isnan(df.loc[index, j]):
                        df.loc[index, "boite"],df.loc[index, j]=df.loc[index, j],df.loc[index, "boite"]
        a=True
swap_boite(data)


data.head()
```

```python
def swap_carburant(df):
    L=['carburant', 'Ville','boite', 'puissance']
    a=True
    for index, row in df.iterrows():
        for i in L:
            if df.loc[index, i] in ["Diesel","Essence","Electrique","LPG","Hybride"]:
                a=False
                df.loc[index, "carburant"],df.loc[index, i]=df.loc[index, i],df.loc[index,"carburant"]
                break;
        if a:
            for j in L:
                if isinstance(df.loc[index, j], float) :
                    if math.isnan(df.loc[index, j]):
                        df.loc[index, "carburant"],df.loc[index, j]=df.loc[index, j],df.loc[index, "carburant"]
        a=True
swap_carburant(data)
data.tail()
```

```python
for index, row in data.iterrows():
    length=len(str(data.loc[index,"price"]))
    if data.loc[index,"price"]!="nan":
        if length>4 and "\u202f" in data.loc[index,"price"] :
            mille=float(data.loc[index,"price"][:length-3].split("\u202f")[0])
            unit=float(data.loc[index,"price"][:length-3].split("\u202f")[1])
            data.loc[index, "price"]=1000*mille+unit
        else:
            data.loc[index, "price"]=math.nan

    else:
        data.loc[index, "price"]=math.nan
data.head()      
```

```python
data[['lower', 'upper']] = data['Kilométrage'].str.split(' - ', expand=True)
data['Kilométrage'] = data['Kilométrage'].replace(['Plus de 500000', 'nan'], np.nan)

# extract the lower and upper limits from the 'Kilométrage' column
data[['lower', 'upper']] = data['Kilométrage'].str.split(' - ', expand=True)

# remove the whitespace and non-digit characters, and replace NaN with NaN
data['lower'] = data['lower'].str.replace(r'\D+', '').replace(np.nan, np.nan).astype(float)
data['upper'] = data['upper'].str.replace(r'\D+', '').replace(np.nan, np.nan).astype(float)

# calculate the mean
data['mean'] = (data['lower'] + data['upper']) / 2
data.head()
```

```python

# replace 'Kilométrage' values with 'mean' values
data.loc[~data['mean'].isna(), 'Kilométrage'] = data['mean']
data = data.drop(columns=['lower', 'upper','mean'])

# display the result
data.head()
```
