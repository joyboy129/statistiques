```python
df1=df.copy()
def swap_kilometrage(df):
    
    L=["Etat","origine", "model","année modele","premiere main","Nombre portes","Marque","boite","carburant","puissance"]
    a=True
    for index, row in df.iterrows():
        for i in L:
            
            if isinstance(df.loc[index, i], str) and "- " in df.loc[index, i]:
                a=False
                df.loc[index, "kilometrage"],df.loc[index, i]=df.loc[index, i],df.loc[index, "kilometrage"]
                break;
            elif df.loc[index, i]=="Plus de 500 000":
                a=False
                df.loc[index, "kilometrage"],df.loc[index, i]=df.loc[index, i],df.loc[index, "kilometrage"]
                break;
             
        if a:
            for j in L:
                if isinstance(df.loc[index, j], float) :
                    if math.isnan(df.loc[index, j]):
                        df.loc[index, "kilometrage"],df.loc[index, j]=df.loc[index, j],df.loc[index, "kilometrage"]
            
        a=True
swap_kilometrage(df1)
df1.head()
```

```python
def swape_annee(df):
    L=["Etat","origine", "model","année modele","premiere main","Nombre portes","Marque","boite","carburant","puissance"]
    a=True
    for index, row in df.iterrows():
        for i in L:
            try:
                if float(df.loc[index, i])<=2024 and 1979<=float(df.loc[index, i]):
                    a=False
                    df.loc[index, "année modele"],df.loc[index, i]=df.loc[index, i],df.loc[index, "année modele"]
                    break;
        
            except ValueError:
                pass
        if a:
            for j in L:
                if isinstance(df.loc[index, j], float) :
                    if math.isnan(df.loc[index, j]):
                        df.loc[index, "année modele"],df.loc[index, j]=df.loc[index, j],df.loc[index, "année modele"]
                
                    
        a=True
            
swape_annee(df1)
df1.head()
```

```python
df1["price"] = df1["price"].astype(str)
for index, row in df1.iterrows():
    length=len(str(df1.loc[index,"price"]))
    if df1.loc[index,"price"]!="nan":
        if length>4 and "\u202f" in df1.loc[index,"price"] :
            mille=float(df1.loc[index,"price"][:length-3].split("\u202f")[0])
            unit=float(df1.loc[index,"price"][:length-3].split("\u202f")[1])
            df1.loc[index, "price"]=1000*mille+unit
        else:
            df1.loc[index, "price"]=math.nan

    else:
        df1.loc[index, "price"]=math.nan
        
df1.head()    
    
```
